```python
df = df.rename(columns={"work_year": "w_year",
                       "experience_level": "ex_level",
                    "employment_type": "emp_type",
                    "job_title": "tittle",
                    "salary": "salary",
                    "salary_currency": "sal_cur",
                    "salary_in_usd": "sal_in_usd",
                    "employee_residence": "emp_res",
                    "remote_ratio": "rem_ratio",
                    "company_location": "com_loc",
                    "company_size" : "com_size"})
```

```python
#количество записей для каждой из профессий
df['tittle'].value_counts()
```

```python
#создадим новый датафрейм с data_analysts
data_analysts = df[df['tittle']=='Data Analyst']
```

```python
data_analysts
```

```python
#посмотрим в компаниях каких стран дата-аналитики в среднем зарабатывают больше всего(+группировка для размера компаниц)
data_an1 = data_analysts.groupby(["com_loc", "com_size"], as_index=False).aggregate({"sal_in_usd":"mean"}).sort_values("sal_in_usd", ascending=False)
```

```python
data_an1
```

```python
#в компании из какой страны и какого размера максимальное значение зарплаты у дата-аналитика,
#какой experience_level у дата-аналитика
top_companyDA = data_analysts.groupby(["com_loc", "com_size","ex_level"], as_index = False)\
    .agg({"sal_in_usd":"max"})\
    .sort_values('sal_in_usd', ascending=False)
```

```python
#максимальная зарпалата у дата-аналитика в США, experience_level - SE
top_companyDA.head(5)
```

```python
#сравним среднею зарплату дата-аналитика и для других топ-5 профессий
top_5 = df[df['tittle'].isin (['Data Scientist','Data Engineer','Data Analyst','Machine Learning Engineer','Research Scientist'])]
```

```python
middle_top_5 = top_5[top_5['ex_level']=='MI']
```

```python
middle_top_5
```

```python
salary_mean_top_5 = middle_top_5.groupby(['tittle'], as_index = False)\
   .aggregate({'sal_in_usd':'mean'}, as_index = False)\
   .sort_values(['sal_in_usd'], ascending = False)
salary_mean_top_5
```

```python
#сравним с "близкими" профессиями дата-аналитику
job_data = df[df['tittle'].isin (['Data Analyst', 'BI Data Analyst', 'Business Data Analyst', 'Data Analytics Engineer', 'Data Analytics Manager'])]
```

```python
middle_job_data = job_data[job_data['ex_level']=='MI']
```

```python
salary_mean_job_data = middle_job_data.groupby(['tittle'], as_index = False)\
   .aggregate({'sal_in_usd':'mean'}, as_index = False)\
   .sort_values(['sal_in_usd'], ascending = False)
salary_mean_job_data
```

```python
#добавим в датафрейм столбец с премией 
for key in data_analysts.ex_level:
    if key == 'SE' or key=='EX':
        data_analysts['bonus_sal']=data_analysts.sal_in_usd*1.10
    else:
        data_analysts['bonus_sal']=data_analysts.sal_in_usd*1.05
data_analysts
```

```python
data_min = data_analysts.groupby(["com_loc", "com_size","ex_level"], as_index = False)\
    .agg({"sal_in_usd":"min"})\
    .sort_values('sal_in_usd')
```

```python
#посмотрим в каких странах и компаниях какого размера самые минимальные зарплаты у дата-аналитика
#какой experience_level дата-аналитика
data_min
```
